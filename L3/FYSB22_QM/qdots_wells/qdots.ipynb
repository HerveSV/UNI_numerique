```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_green = pd.read_csv("data/emission/green_spectrum.csv", delimiter=";", decimal=",")
data_orange = pd.read_csv("data/emission/orange_spectrum.csv", delimiter=";", decimal=",")
data_reference = pd.read_csv("data/emission/reference.csv", delimiter=";", decimal=",")
```

```python
wavelength_green = data_green["Wavelength (nm)"].to_numpy()
I_green = data_green["Intensity green"].to_numpy()

wavelength_orange = data_orange["Wavelength (nm)"].to_numpy()
I_orange = data_orange["Intensity orange"].to_numpy()

wavelength_ref = data_reference["Wavelength (nm)"].to_numpy()
I_ref = data_reference["Intensity reference"].to_numpy()

print(wavelength_green)
```

```python
def get_max_index(X):
    max_index = 0
    for i in range(len(X)):
        if X[i] > X[max_index]:
            max_index = i

    return max_index
```

```python
green_lambda_max = wavelength_green[get_max_index(I_green-I_ref)]

plt.plot(wavelength_green, I_green-I_ref)
plt.axvline(green_lambda_max, c="red")
plt.show()

print(f"{green_lambda_max=:.3} nm")

orange_lambda_max = wavelength_orange[get_max_index(I_orange-I_ref)]

plt.plot(wavelength_orange, I_orange-I_ref)
plt.axvline(orange_lambda_max, c="red")
plt.show()

print(f"{orange_lambda_max=:.3} nm")
```
