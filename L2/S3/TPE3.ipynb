```python
import numpy as np
import matplotlib.pyplot as plt
```

## Cas 1 R = 100 $Î©$

```python
L = 100e-3
C = 100e-9

R = 100

def gain(f, R, L, C):
    omeg = 2*np.pi*f
    Q = 1/R*np.sqrt(L/C)
    w0 = 1/np.sqrt(L*C)
    denom = np.sqrt(1+Q**2*(omeg/w0 - w0/omeg)**2)

    return 1/denom

def phase(f, R, L, C):
    omeg = 2*np.pi*f
    Q = 1/R*np.sqrt(L/C)
    w0 = 1/np.sqrt(L*C)

    return -np.arctan(Q*(omeg/w0 - w0/omeg))

def db(X):
    return 20*np.log10(X)

f_theo = np.array([50, 100, 250, 500, 1000, 1200, 1400, 1500, 1600, 1700, 1800, 2000, 2300, 2800, 4000, 6000, 8000, 1e4, 3e4, 5e4])
f = np.linspace(20, 5e4, 10000)

G = gain(f, R, L, C)
G_db = db(G)

G_theo = gain(f_theo, R, L, C)
G_db_theo = db(G_theo)

phi = phase(f, R, L, C)
phi_theo = phase(f_theo, R, L, C)



plt.xscale("log")

plt.plot(f, G)
plt.scatter(f_theo, G_theo, c="r", zorder=4)

f0 = 1/np.sqrt(L*C)/(2*np.pi)
print(f0)

plt.axvline(f0)

plt.show()




plt.xscale("log")

plt.plot(f, G_db)
plt.scatter(f_theo, G_db_theo, c="r", zorder=4)

plt.axvline(f0)

plt.show()


plt.xscale("log")

plt.plot(f, phi)
plt.scatter(f_theo, phi_theo, c="r", zorder=4)

plt.axvline(f0)

plt.show()



#for i in range()
    

```

```python
for i in range(len(f_theo)):
    print(f"f = {f_theo[i]}")
    print(f"gain = {G_theo[i]}")
    print(f"gain_db = {G_db_theo[i]}")
    print(f"phi = {phi_theo[i]}")
    print()
```
