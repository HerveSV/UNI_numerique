```python
import matplotlib.pyplot as plt
import numpy as np
```

## Additivit√© des gains

```python
w1 = 100
w2 = 1000

w = np.linspace(1, 1e4, 10000)

H2 = (w/w2)*1j
H1 = 1/(1+(w/w1)*1j)

H3 = 10*H1*H2

G1 = np.abs(H1)
G2 = np.abs(H2)
G3 = np.abs(H3)

G1_db = -10 * np.log10(1+(w/w1)**2)
#20*np.log(G1)
G2_db = 20*np.log10(w) - 20*np.log10(w2)
#20*np.log(G2)

G3_db = 20*np.log10(G3)

plt.xscale("log")
plt.plot(w, G1_db, label="G1_db")
plt.plot(w, G2_db, label="G2_db")

plt.axhline(0, c="black")

plt.plot(w, G3_db, label="G3_db = 20log(10*G1*G2)")

plt.legend()
plt.show()
```

```python
w0 = 10

H4 = (1 + (w/(10*w0)*1j))
H5 = (1 + (w/(40*w0)*1j))

H6 = 1/(1 + (w/(w0)*1j))
H7 = 1/(1 + (w/(100*w0)*1j))

H8 = H4*H5*H6*H7

G4_db = 20*np.log10(np.abs(H4))
G5_db = 20*np.log10(np.abs(H5))
G6_db = 20*np.log10(np.abs(H6))
G7_db = 20*np.log10(np.abs(H7))

G8_db = 20*np.log10(np.abs(H8))


plt.xscale("log")
plt.plot(w, G4_db, label="G4_db")
plt.plot(w, G5_db, label="G2_db")
plt.plot(w, G6_db, label="G6_db")
plt.plot(w, G7_db, label="G7_db")

plt.axhline(0, c="black")
plt.axhline(-12, c="black")
plt.axhline(-20, c="black")


plt.plot(w, G4_db+G5_db+G6_db+G7_db, label="G8_db")

plt.plot(w, G8_db, label="G8_db")

plt.legend()

plt.show()


plt.xscale("log")
plt.plot(w, np.angle(H4), label="G4_db")
plt.plot(w, np.angle(H5), label="G4_db")
plt.plot(w, np.angle(H6), label="G4_db")
plt.plot(w, np.angle(H7), label="G4_db")




plt.plot(w, np.angle(H8), label="G4_db")



plt.legend()
plt.show()
```
