```python
filepath = "/Users/hervesv/Documents/Stuff/Projects/UNI_numerique/L2/TP_vib_1/fichier_python/data/TP_pendule-20231002-HSV&PT/petits_angles/TP_Pendule_PA_2023-10-02_14-47-02_final-0.25kg.csv"
data = pd.read_csv(filepath, header=None)

#print(data)

print("yyeeeet")


data_n = data.transpose().to_numpy()
#print(data_n)

measures = [[[],[]] for i in range(5)]
c = 0


for i in range(len(measures)):
    for j in range(len(measures[i])):
        measures[i][j] = data_n[c]
        c += 1

#print(measures[1])

""" 2) Determine number of data points """

#datafile = open(filename, 'r')
#ndata = 0
#for line in datafile:
#    ndata = ndata + 1
#print("Number of data points =",ndata)
#datafile.close

""" 3) Fill in data array """

#data = np.zeros((ndata,2))

#datafile = open(sys.argv[1], 'r')
#ndata = 0
#for line in datafile:
#    words = line.split(',')
#    data[ndata,0] = words[0]
#    data[ndata,1] = words[1]
#    ndata = ndata + 1
#datafile.close

""" 4) Fit curve """

def sinusoide_amortie(t,amplitude,phase,period,damping,offset):
    return amplitude*cos(2*pi*t/period-phase)*exp(-t/damping)+offset

def get_index_from_time(t,tab):
    if t < min(tab) or t > max(tab):
        return len(tab)-1
    else:
        for i in range(len(tab)-2):
            if t > tab[i] and t < tab[i+1]:
                return i
        return len(tab)-1
    

def calc_OC(R, l, h, m, mt, ma, n):
    return (m*(R+l+0.5*n*h)+mt*(R+l/2))/(mt+m+ma)

def calc_IOx(R, l, h, radius, m, mt, ma, n):
    return m*(R+l+n*h/2)**2+0.25*m*(radius**2+(n*h)**2/3) + mt*l**2/12+mt*(R+l/2)**2+0.5*ma*R**2

def calc_T0(IOx, OC, g, m, mt, ma):
    return 2*pi*sqrt(IOx/((m+mt+ma)*g*OC))

#theta_T = [np.zeros(2,1) for i in range(len(measures))]
theta0s250 = np.zeros(len(measures))
Ts250 = np.zeros(len(measures))
sig_Ts250 = np.zeros(len(measures))

for j in range(len(measures)):
      tmax = max(measures[j][0])
      #i = get_index_from_time(tmax,data[:,0])
      """print("Time t =",tmax,"Index i =",i)"""
      times = measures[j][0]
      angles = measures[j][1]
      params,cov = curve_fit(sinusoide_amortie,times,angles)
      t = linspace(0,tmax, 1000)
      (amplitude,phase,period,damping,offset) = params
      
      
      plt.plot(times,angles, 'o', color="r")
      plt.xlabel(r"$\theta$ (rad)")
      plt.xlabel("$t$ (s)")
      plt.title(rf"volution de $\theta$ en fonction de $t$, $\theta_0$={np.rad2deg(amplitude):.1f} deg.")
      
      plot(t,sinusoide_amortie(t,amplitude,phase,period,damping,offset), '-', zorder=-2)
      show()
      
      print("Fit over time interval [0;",tmax,"]")
      print("Amplitude theta0 =",'{:.3f}'.format(amplitude),"rad =",
            '{:.1f}'.format(amplitude*180/pi),"deg")
      print("Period T         =",'{:.3f}'.format(period),"s")
      print("Damping time     =",'{:.0f}'.format(damping),"s")
      print("Offset angle     =",'{:.3f}'.format(offset),"rad =",
            '{:.1f}'.format(offset*180/pi),"deg")

      """ 5) Theory """

      g = 9.81

      R = 0.032   # R = 3.2 cm
      sig_R = 0.002
      u_R = ufloat(R, sig_R)

      mc = 0.277 # kg
      sig_mc = 0.001
      u_mc = ufloat(mc, sig_mc)

      radius = 0.025 # r = 2.5 cm
      sig_radius = 0.001
      u_radius = ufloat(radius, sig_radius)
      
      h = 0.017      # h = 1.7 cm
      sig_h = 0.001
      u_h = ufloat(h, sig_h)
      
      n = 1
      
      m = n*mc
      sig_m = sig_mc*n
      u_m = n*u_mc
      
      mt = 0.24 #0.20
      sig_mt = 0.01
      u_mt = ufloat(mt, sig_mt)
      
      ma = 0.4 #0.565
      sig_ma = 0.01
      u_ma = ufloat(ma, sig_ma)

      l = 0.58 # 0.66-(R+n*h/2)  # l = 58 cm
      sig_l = 0.001
      u_l = ufloat(l, sig_l)


      OC = calc_OC(R, l, h, m, mt, ma, n)#(m*(R+l+0.5*n*h)+mt*(R+l/2))/(mt+m+ma)
      IOx = calc_IOx(R, l, h, radius, m, mt, ma, n)#m*(R+l+n*h/2)**2+0.25*m*(radius**2+(n*h)**2/3) + mt*l**2/12+mt*(R+l/2)**2+0.5*ma*R**2
      T0 = calc_T0(IOx, OC, g, m, mt, ma)#2*pi*np.sqrt(IOx/((m+mt+ma)*g*OC))

      max_OC = calc_OC(R+sig_R, l+sig_l, h+sig_h, m+sig_m, mt+sig_mt, ma+sig_ma, n)
      min_OC = calc_OC(R-sig_R, l-sig_l, h-sig_h, m-sig_m, mt-sig_mt, ma-sig_ma, n)
      sig_OC = np.max([np.abs(OC-max_OC), np.abs(OC-min_OC)])

      max_IOx = calc_IOx(R+sig_R, l+sig_l, h+sig_h, radius+sig_radius, m+sig_m, mt+sig_mt, ma+sig_ma, n)
      min_IOx = calc_IOx(R-sig_R, l-sig_l, h-sig_h, radius-sig_radius, m-sig_m, mt-sig_mt, ma-sig_ma, n)
      sig_IOx = np.max([np.abs(IOx-max_IOx), np.abs(IOx-min_IOx)])

      max_T0 = calc_T0(IOx+sig_IOx, OC-sig_OC, g, m-sig_m, mt-sig_mt, ma-sig_ma)
      min_T0 = calc_T0(IOx-sig_IOx, OC+sig_OC, g, m+sig_m, mt+sig_mt, ma+sig_ma)
      sig_T0 = np.max([np.abs(T0-max_T0), np.abs(T0-min_T0)])


      #u_OC = (u_m*(u_R+u_l+0.5*n*u_h)+u_mt*(u_R+u_l/2))/(u_mt+u_m+u_ma)
      #u_IOx = u_m*(u_R+l+n*u_h/2)**2+0.25*u_m*(u_radius**2+(n*u_h)**2/3) + u_mt*u_l**2/12+u_mt*(u_R+u_l/2)**2+0.5*u_ma*u_R**2
      #u_T0 = 2*pi*np.sqrt(u_IOx/((u_m+u_mt+u_ma)*g*u_OC))




      print("\nLength of the rod:   l =",'{:.1f}'.format(l*100),"cm")
      print("Mass of the rod:    mt =",'{:.0f}'.format(mt*1000),"g")
      print("Mass of axial disc: ma =",'{:.0f}'.format(ma*1000),"g")
      print("Total mass: M =",'{:.0f}'.format((m+ma+mt)*1000),"g")
      print("Center-of-mass distance from rotation axis: OC =",
            '{:.1f}'.format(OC*100),"cm")
      print("Harmonic period T0 =",'{:.3f}'.format(T0),"s")
      theta0 = np.abs(amplitude)
      TBorda = T0*(1+theta0**2/16)
      Texact = T0*(1+0.25*sin(theta0/2)**2)

      max_Texact = (T0+sig_T0)*(1+0.25*sin(theta0/2)**2)
      min_Texact = (T0-sig_T0)*(1+0.25*sin(theta0/2)**2)
      sig_Texact = np.max([np.abs(Texact-max_Texact), np.abs(Texact-min_Texact)])


      #u_Texact = u_T0*(1+0.25*sin(theta0/2)**2)
      print("Borda formula   T  =",'{:.3f}'.format(TBorda),
            "s  (with theta0 =",'{:.3f}'.format(theta0),"rad)")
      print("Exact formula   T  =",'{:.3f}'.format(Texact),
            "s  (with theta0 =",'{:.3f}'.format(theta0),"rad)")

      T0=period/(1+theta0**2/16)
      print("\nExperimentally deduced T0 =",'{:.3f}'.format(T0),"s")
      omega0=2*pi/T0
      alpha = l**2/12+(R+l/2)**2
      beta = m*(R+l+n*h/2)**2+ma*R**2/2
      gamma = R+l/2
      delta = m*(R+l+n*h/2)
      mt = (delta-omega0**2/g*beta)/(omega0**2/g*alpha-gamma)
      print("Experimentally deduced mass of the rod:   mt =",
            '{:.0f}'.format(mt*1000),"g")

      rho = mc/(pi*radius**2*h-pi*(0.5*0.007)**2*h)
      #rho = mc/(pi*radius**2*h) # hole neglected
      print("Estimated density of the rod: rho =",
            '{:.0f}'.format(rho),"kg/m^3")
      diameter = 0.0075 # Diameter of the rod (effective solid rod)
      mt = rho*pi*diameter**2/4*l
      print("Estimated mass of the rod (from density): mt =",
            '{:.0f}'.format(mt*1000),"g")
      
      theta0s250[j] = theta0
      Ts250[j] = Texact

      sig_Ts250[j] = sig_Texact

      print(Ts250)




      #quit()
```
